```python
# DATA LOAD
df_clean = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/*.csv")
```

```python
df = spark.read.option("header", True).option("inferSchema", True) \
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/*.csv")

df.show()
df.printSchema()
```

```python
### inner join - returns matching rows from both the tables 
```

```python
#Creating unique users table 

users=df.select("user_id","user_session").distinct()
```

```python
joined_df=df.join(users,"user_id","inner")
joined_df.show(5)
```

```python
### Left join 

joined_df=df.join(users,"user_id","left")
joined_df.show(5)
```

```python
### right join 

joined_df=df.join(users,"user_id","right")
joined_df.show(5)
```

```python
### full join 

joined_df=df.join(users,"user_id","full")
joined_df.show(5)
```

```python
### window function running speed 

from pyspark.sql import Window
from pyspark.sql import functions as F

window_user = Window \
    .partitionBy("user_id") \
    .orderBy("event_time") \
    .rowsBetween(Window.unboundedPreceding, Window.currentRow)

df = df.withColumn(
    "running_spend",
    F.sum("price").over(window_user)
)

df.select(
    "user_id",
    "event_time",
    "price",
    "running_spend"
).show(truncate=False)
```

```python
## window function order sequence 

from pyspark.sql import Window
from pyspark.sql import functions as F

window_user = Window \
    .partitionBy("user_id") \
    .orderBy("event_time")

df = df.withColumn(
    "order_sequence",
    F.row_number().over(window_user)
)

df.select("user_id","event_time","price","order_sequence").show(truncate=False)                                                  
```

```python
## window function ranking

from pyspark.sql import Window
from pyspark.sql import functions as F

window_user = Window \
    .partitionBy("user_id") \
    .orderBy("event_time")

df = df.withColumn(
    "rank",
    F.rank().over(window_user)
)

df.select(
    "user_id",
    "event_time",
    "price",
    "rank"
).orderBy("user_id", "event_time").show(truncate=False)
```

```python
## window function category ranking 
```

```python
from pyspark.sql import Window
from pyspark.sql import functions as F

window_category = Window \
    .partitionBy("category_code") \
    .orderBy("price")

df_ranked = df_clean.withColumn(
    "price_rank",
    F.rank().over(window_category)
)

df_ranked.select("category_code","product_id","brand","price","price_rank").orderBy("category_code", "price_rank") \
 .show(truncate=False)
```
