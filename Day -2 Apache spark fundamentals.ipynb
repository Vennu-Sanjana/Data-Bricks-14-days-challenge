```python
df = spark.read.option("header", True).option("inferSchema", True) \
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/*.csv")
```

```python
## performing Basic Operations
```

```python
## Select command - Selecting specific columns 

df.select("event_type", "product_id","price").show(5)
```

```python
## Finding High value products - Filter function 

df.filter(df.price > 1000.00).show(5)
```

```python
# Grouping function 

df.groupBy("category_code").count().show(5)
```

```python
df.groupBy("event_type","category_code").count().show(5)
```

```python
### sorting data - order by 

df.orderBy(df.price.desc()).show(5)
```

```python
### finding distinct records 

df.select("brand").distinct().show(5)
```

```python
df.select('event_type').distinct().show(5)
```

```python
### exporting the values and saving it as a table 

df.write \
    .mode("overwrite") \
    .saveAsTable("default.my_output_table")
```
